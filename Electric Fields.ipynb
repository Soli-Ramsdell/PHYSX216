```python
import physplot as phys
import numpy as np

#Measurements
deltaR = np.array([0.300, 0.304, 0.298, 0.297]) #m
deltaV = np.array([3.3, 8.1, 16.8, 42.1]) #V
rMid = np.array([0.251, 0.553, 0.854, 1.153]) #m

#Calculations
E = deltaV / deltaR
logE = np.log(E)
logRMid = np.log(rMid)

#Error analysis
errR = np.ones(4)*0.005
errV = np.ones(4)*0.5
errDeltaR = np.sqrt(errR**2+errR**2) #Rule 3
errDeltaV = np.sqrt(errV**2+errV**2) #Rule 3
errE = E * np.sqrt((errDeltaR/deltaR)*(errDeltaR/deltaR) + (errDeltaV/deltaV)*(errDeltaV/deltaV)) #Rule 4
errLogE = np.log(E + errE) - np.log(E) #Brute force


#plotfit takes in parameters in this order
#phys.plotfit(x,y,dy,title,'xlabel(units)','ylabel(units)','slope units')  Slope units will not render in LaTeX

phys.plotfit(logRMid,logE,errLogE,'Electric Fields','ln(rMid)($m$)','ln(E)($V$)','V/m')
```
