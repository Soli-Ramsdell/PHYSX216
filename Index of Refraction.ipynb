```python
#Soli Ramsdell
#PHYSX 218
#Refraction
#15 April 2021

import physplot as phys
import numpy as np

thetaI = np.deg2rad([10,15,20,25,30,35,40,45,50,60])#deg
thetaR = np.deg2rad([7,11,15,19,22,26,29,32,35,45])#deg
sinThetaI = np.sin(thetaI)#deg
sinThetaR = np.sin(thetaR)#deg

errSinTheta = np.sin(thetaI+0.5)-np.sin(thetaI)

#plotfit takes in parameters in this order
#phys.plotfit(x,y,dy,title,'xlabel(units)','ylabel(units)','slope units')  Slope units will not render in LaTeX

phys.plotfit(sinThetaR,sinThetaI,errSinTheta,'Index of Refraction of Water','sin(thetaI)($deg$)','sin(thetaR)($deg$)','')
```
